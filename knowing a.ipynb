```python
EV= .6
X = arange(.01,1-EV,.01)
style.use("fivethirtyeight")
scatter(X, (log(.4)-log(X))/log(600))
title("$b$ vs $a$ when\n$EV[r]=%.1f,M=600$"%EV)
xticks([0,.2,.4]); yticks([0,.2,.4,.6])

def f(x,a,b):
    return a*x**b
```

```python
def evansf(x,a): return a*x**((log(.3) - log(a)) / log(1200))

figsize(4,4)
k=1
for aa in arange(0.02,.102,.02):
    #subplot(510+k)
    #XX = arange(0,.1,.02)
    XX = range(1200)
    plot(XX,evansf(XX,aa))
    #title("a:"+str(aa))
    k+=1
```

Above: plot gets steeper as $a$ gets larger

```python
evansf(200,.02)
```
